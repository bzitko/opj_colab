```python
import re
from collections import defaultdict
from nltk.corpus import wordnet as wn
from nltk import word_tokenize
```

```python
filename = "tweet_global_warming.csv"

dataset = []
classes = set()
for line in open(filename, encoding="utf8").read().strip().split("\n")[1:]:
    *txt, cls, conf = line.split(",")
    txt = "".join(txt).strip()
    if cls in ("Y", "Yes"):
        cls = "Yes"
    elif cls in ("N", "No"):
        cls = "No"
    else:
        cls = "Maybe"
        
    dataset.append((txt, cls))
    classes.add(cls)
```

```python
cleandataset = [(" ".join([w for w in word_tokenize(txt) if wn.synsets(w)]), cls) for txt, cls in dataset]
cleandataset
```

```python
patterns = [
#            r"\[.*?\]",
            r"\[link\]",
            r"\[pic.*?\]",
            r"\[PIC.*?\]",    
           ]

samples = defaultdict(int)
for txt, _ in dataset:
    for pattern in patterns:
        for m in re.findall(pattern, txt):
            samples[m] += 1

samples
```
